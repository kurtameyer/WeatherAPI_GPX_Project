```python
import pandas as pd
import requests
import datetime
import json
```

```python
#Initial test data to learn API 

API_KEY = '!?!?!?!?!?!?'
DATE = '2023-06-22 16:30'
LATITUDE = '40.0203360'
LONGITUDE = '-79.5962450'
```

```python
#Construct URL 
HISTORICAL_DATA_URL = f"http://api.weatherapi.com/v1/history.json?key={API_KEY}&q={LATITUDE},{LONGITUDE}&dt={DATE}"
```

```python
#Response

response = requests.get(HISTORICAL_DATA_URL)
```

```python
#Inspect JSON 
data = response.json()
print(json.dumps(data, indent=4)) 
#You're going to have to parse name, region, and localtime out of the first dictionary
#Times aren't always going to match calls 
#Second dictionary you have lots of nested dictionaries. Pull temp_f, condition, wind_mph, humidity out
```

```python
for hour_data in data["forecast"]["forecastday"][0]["hour"]:
    temp_f = hour_data["temp_f"]
    hourly = hour_data["time"]
    condition_text = hour_data["condition"]["text"]
    #print data
    print(f"Time: {hourly}, Temperature: {temp_f}°F, Condition: {condition_text}")
```
